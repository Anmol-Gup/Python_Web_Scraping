# ICC Test Rankings

```python
def get_columns(table):
    return test_table.select('th')
```

```python
def get_data(table,team_ranking):
    rows=table.find_all('tr')
    for row in rows[1:]:
        if row.text!='\n':
            #print(i.text)
            s=row.text.strip().split('\n')
            team_ranking['Pos'].append(s[0])
            team_ranking['Team'].append(s[1])
            team_ranking['Matches'].append(s[2])
            team_ranking['Points'].append(s[3])
            team_ranking['Rating'].append(s[4])
    return team_ranking
```

```python
tables=soup.select('.StoryengineTable')
test_table=tables[0]

columns=get_columns(test_table)
test_team_rankings={columns[0].text:[],columns[1].text:[],columns[2].text:[],columns[3].text:[],columns[4].text:[]}

test_team_rankings=get_data(test_table,test_team_rankings)
test_team_rankings_df=pd.DataFrame(test_team_rankings)
test_team_rankings_df
```
