# List of Notable companies in India

```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup
import json
```

```python
r=req.get('https://en.wikipedia.org/wiki/List_of_companies_of_India')
soup=BeautifulSoup(r.content,'html.parser')
soup.prettify()
```

```python
companies_table=soup.select('table')[1]
head=companies_table.select('th')
companies=[]

for row in companies_table.find_all('tr')[1:]:
    company=row.find_all('td')
    companies.append({'Name':company[0].get_text()[:-1],'Industry':company[1].get_text()[:-1],'Sector':company[2].get_text()[:-1],
                         'Headquarters':company[3].get_text()[:-1],'Founded':company[4].get_text()[:-1],'Notes':company[5].get_text()[:-1]})

companies
with open('companies.json','w') as file:
    json.dump(companies,file)
```
